```python
import pandas as pd
import numpy as np

# =========== q1
b_sum = """
2.109 2.086 2.036 2.038 2.015 2.014 2.005 2.003 1.984 1.981 
1.963 1.957 1.908 1.894 2.066 2.075 2.031 2.029 2.013 2.014 
1.999 1.996 1.973 1.975 1.951 1.951 2.065 2.057 2.025 2.029 
2.012 2.012 1.997 1.992 1.971 1.969 1.947 1.941 2.052 2.044 
2.023 2.020 2.012 2.010 1.994 1.986 1.966 1.967 1.941 1.938
"""
b_sum = b_sum.split()
```

```python
# H0: u ==  2.0 litres
# H1: u != 2.0 litres
df = pd.DataFrame(b_sum)
```

```python
x = np.array(df)
y = x.astype(np.float)
```

```python
print(len(y))
```

```python
y_len = len(y)

x_mean = np.mean(y)
x_stdiv = np.std(y)

print("Mean =====> ",x_mean) 
print("Std div ===========> ",x_stdiv, " \n ")


# z-score

z_score = ( (x_mean - 2.0) / ( (x_stdiv) / (np.sqrt(y_len) )))

print("Z-SCORE ======================> ",z_score)

```

```python
# p-value

# conf_level = 0.05

from scipy.stats import norm
prob = norm.cdf(z_score)
print(prob * 2)

# ============== conclusion ==========
```
