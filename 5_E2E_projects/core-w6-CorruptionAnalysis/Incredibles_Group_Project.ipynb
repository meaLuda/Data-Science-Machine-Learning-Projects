### Functions for data cleaning 

```python
# Creating a function to load our dataset and preview the first 3 rows 

def load_dataset(data):
  df=pd.read_csv(data, index_col='No.')
  return df


#Checking for duplicates 

def duplicates(val):
  no_null=val.isnull().sum()
  return no_null

# Function to save our cleaned data in form of CSV and XLSX

def save_data_set(data, name):
  data.to_csv(name)
  return data 

#Function to check information on a dataset 

def check_info(val):
  dataframe_info= val.info()
  return dataframe_info

# A function to strip white space Stripping space 
def strip_space(val):
  stripped_data=val.str.strip()
  return stripped_data

# A function to drop null values 
def drop_null_values(value):
  not_null=value.dropna()

  return not_null
```
