## Q1: How many visits did Kimathi Street and Pipeline medical centers' have from May 2022 and September 2022?

```python
df_visit.head()
```

```python

# Convert VisitDateTime column to datetime type
df_visit['VisitDateTime'] = pd.to_datetime(df_visit['VisitDateTime'])

# Filter visits from May 2022 to September 2022 and for the specified medical centers
start_date = pd.Timestamp('2022-05-01')
end_date = pd.Timestamp('2022-09-30')
medical_centers = ['Kimathi Street', 'Pipeline']

filtered_df = df_visit[
    (df_visit['VisitDateTime'] >= start_date) &
    (df_visit['VisitDateTime'] <= end_date) &
    (df_visit['MedicalCenter'].isin(medical_centers))
]

# Count the number of visits
visit_count = len(filtered_df)

print("Number of visits to Kimathi Street and Pipeline medical centers between May 2022 and September 2022:", visit_count)
```

# 2. What was the most common diagnosis in 2022 for Tassia and Embakasi branches combined?

```python
df_visit.head()
```

```python
df_Diagnosis.head()
```

```python
# Merge the two dataframes based on the VisitCode column
merged_df = pd.merge(df_visit, df_Diagnosis, on="VisitCode")

# Filter the data for the year 2022
merged_df["VisitDateTime"] = pd.to_datetime(merged_df["VisitDateTime"])
merged_df_2022 = merged_df[merged_df["VisitDateTime"].dt.year == 2022]

# Count the occurrences of each diagnosis
diagnosis_counts = merged_df_2022["Diagnosis"].value_counts()

# Get the most common diagnosis
most_common_diagnosis = diagnosis_counts.idxmax()

# Print the result
print("The most common diagnosis in 2022 for Tassia and Embakasi branches combined was:", most_common_diagnosis)
```

# 3. Which payor was the most profitable (in absolute numbers) for Penda Health in 2022? {Assume a gross average margin of 30% per visit.


```python
df_visit.head()
```

```python
df_Invoice.head()
```

```python
# Merge the two dataframes
df_profit = df_visit.merge(df_Invoice, on='VisitCode')

# Calculate the gross profit per visit
df_profit['GrossProfit'] = df_profit['Amount'] * 0.3

# Calculate the total gross profit for each payor
payor_profit = df_profit.groupby('Payor')['GrossProfit'].sum()

# Print the payor with the highest total gross profit
print("The payor with the highest total gross profit:",payor_profit.sort_values(ascending=False).index[0])
```

# 4. Which medical center was the least profitable  (in absolute numbers)  in 2022?  {Assume a gross average margin of 30% per visit}.

```python
df_visit.head()
```

```python
df_Invoice.head()
```

```python
# Merge the two dataframes
df_merged = df_visit.merge(df_Invoice, on='VisitCode')

# Calculate the gross profit per visit
df_merged['GrossProfit'] = df_merged['Amount'] * 0.3

# Calculate the total gross profit for each medical center
df_merged = df_merged.groupby('MedicalCenter')['GrossProfit'].sum()

# Find the medical center with the lowest total gross profit
least_profitable_medical_center = df_merged.sort_values().index[0]

print(f"The least profitable medical center in 2022 was: {least_profitable_medical_center}")
```

# 8. In 2022 what proportion of visits in Penda Health were second visits? *{A first visit is when a patient visits Penda Health for the very first time, a second visit is when the same patient visits again i.e. a second time and so on}*


```python
df_visit.head()
```

```python
# Get the year of each visit
df_visit['VisitYear'] = df_visit['VisitDateTime'].dt.year

# Calculate the number of first visits and second visits in 2022
first_visits_2022 = len(df_visit.loc[(df_visit['VisitYear'] == 2022) & (df_visit['PatientCode'].duplicated() == False)])
second_visits_2022 = len(df_visit.loc[(df_visit['VisitYear'] == 2022) & (df_visit['PatientCode'].duplicated() == True)])

# Calculate the proportion of second visits in 2022
proportion_second_visits_2022 = second_visits_2022 / (first_visits_2022 + second_visits_2022)

print(f"The proportion of second visits in 2022 is {proportion_second_visits_2022:.2%}")
```

# 9. What percent of visits in April 2022 happened within 30 days of the preceding visit by the same patient?

```python
# Get the year and month of each visit
# df_visit['VisitYear'] = df_visit['VisitDateTime'].dt.year
df_visit['VisitMonth'] = df_visit['VisitDateTime'].dt.month

# Calculate the difference in days between each visit and the preceding visit
df_visit['DaysSincePriorVisit'] = (df_visit['VisitDateTime'] - df_visit['VisitDateTime'].shift(1)).dt.days

# Filter the dataframe to only include visits in April 2022
df_visit_april = df_visit.loc[(df_visit['VisitYear'] == 2022) & (df_visit['VisitMonth'] == 4)]

# Calculate the number of visits within 30 days of the preceding visit
visits_within_30_days = len(df_visit_april.loc[df_visit_april['DaysSincePriorVisit'] <= 30])

# Calculate the percentage of visits within 30 days of the preceding visit
percentage_visits_within_30_days = visits_within_30_days / len(df_visit_april) * 100

print(f"The percentage of visits in April 2022 that happened within 30 days of the preceding visit by the same patient is {percentage_visits_within_30_days:.2f}%")
```
