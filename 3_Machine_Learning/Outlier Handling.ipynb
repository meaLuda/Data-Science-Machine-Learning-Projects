```
# Loading our libraries
# ---
#  
import pandas as pd
import numpy as np
```

```
# Creating a DataFrame
# 
houses = pd.DataFrame()
houses['Price'] = [534433, 392333, 293222, 4322032]
houses['Bathrooms'] = [2, 3.5, 2, 116]
houses['Square_Feet'] = [1500, 2500, 1500, 48000]
houses
```

```
# Option 1: Drop observations greater than some value
# ---
# 
houses[houses['Bathrooms'] < 20]
```

```
# Option 2: Marking the observations. Create featuring based on boolean condition
# ---
# 
houses['Outlier'] = np.where(houses['Bathrooms'] < 20, 0, 1)

# Show data
houses
```

```
# Option 3: Rescaling: Log values so that outliers don't have as great an effect.
# ---
# 
houses['Log_Of_Square_Feet'] = [np.log(x) for x in houses['Square_Feet']]

# Show data
houses
```
